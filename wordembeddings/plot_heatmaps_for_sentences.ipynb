```python
cosine_distances = []

for outer_template_vector in utah_cleaned:
    temp_cosine_distances = []
    for inner_template_vector in utah_cleaned:
        temp_cosine_distances.append(cosine(outer_template_vector, inner_template_vector))
    cosine_distances.append(temp_cosine_distances)
    
fig = go.Figure(data=go.Heatmap(
                    z=cosine_distances))
fig.update_layout(
    title="With cleaning"
)
fig.show()
```

```python
##################################### CLEANING ####################################
cosine_distances_with_cleaning = []
for outer_template_vector in utah:
    temp_cosine_distances = []
    for inner_template_vector in utah:
        temp_cosine_distances.append(cosine(outer_template_vector, inner_template_vector))
    cosine_distances_with_cleaning.append(temp_cosine_distances)
    
fig2 = go.Figure(data=go.Heatmap(
                    z=cosine_distances_with_cleaning))
fig2.update_layout(
    title="Without cleaning"
)
fig2.show()

#for sent in enumerate(cosine_distances):
#    for a in sent:
#        print(a)
```

```python
cosine_distances_finetune = []

for outer_template_vector in utah_finetune:
    temp_cosine_distances = []
    for inner_template_vector in utah_finetune:
        temp_cosine_distances.append(cosine(outer_template_vector, inner_template_vector))
    cosine_distances_finetune.append(temp_cosine_distances)
    
fig = go.Figure(data=go.Heatmap(
                    z=cosine_distances_finetune))
fig.show()
```

```python
import numpy as np

mean_difference = []
std_dev = []

for block in cosine_distances_with_cleaning:
    tmp = []
    for val in block:
        if val != 0.0:
            tmp.append(val)
    mean_difference.append(np.mean(tmp))
    std_dev.append(np.std(tmp))

print("mean:")    
for difference in mean_difference:
    print(difference)
print("---------")
print("std:")
for std in std_dev:
    print(std)
```
